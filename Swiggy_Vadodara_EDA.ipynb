```python
#@title
from bs4 import BeautifulSoup
import requests
import pandas as pd 
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

url='https://www.swiggy.com/city/Vadodara'

# Creating empty list for features.

```python
#@title
# creating empty list for features.
name=[]
type_=[]
rating = []
time=[]
amount =[]
for page in range(1,30):
    url = 'https://www.swiggy.com/vadodara?page={}'.format(page)
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    for res_name in soup.find_all('div',class_ = "nA6kb"):
        name.append(res_name.text)
    for res_time in soup.find_all('div',class_ = '_3Mn31'):
        time.append(res_time.text)
    for i in soup.find_all('div',class_='_1gURR'):
        type_.append(i.text)
```

# Biryani's

```python
#@title
d = {'Restaurant':name,
    'Type':type_,
    
    'Time':time}
```

```python
#@title
df= pd.DataFrame(d)
```

```python
#@title
len(type_)
```

# CSV IMPORT

```python
#@title
df.to_csv('swiggy_biryani.csv')
```
